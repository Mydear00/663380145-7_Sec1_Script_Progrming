ให้เลือกว่าจะใช้ List หรือ Tuple พร้อมเหตุผล

| สถานการณ์                       | ควรใช้    | เหตุผล                        |
| ------------------------------- | --------- | ----------------------------- |
| 1. อุณหภูมิรายวัน 1 สัปดาห์     | **List**  | สามารถเพิ่มหรือแก้ไขข้อมูลได้ |
| 2. ค่า RGB ของสี เช่น (255,0,0) | **Tuple** | ค่าไม่ควรเปลี่ยนแปลง          |
| 3. สินค้าในตะกร้าสินค้า         | **List**  | เพิ่มหรือลบสินค้าได้          |
| 4. วันในสัปดาห์                 | **Tuple** | เป็นข้อมูลคงที่               |
| 5. คะแนนนักเรียน                | **List**  | คะแนนอาจมีการแก้ไข            |
| 6. พิกัด GPS ของสถานที่         | **Tuple** | พิกัดเป็นค่าคงที่             |
